```python
#import some important libraries
import numpy as np
import pandas as pd

#now read the data using pandas
dataset = pd.read_csv('seattle-weather.csv')
```

```python
dataset
```

```python
final_dataset = pd.get_dummies(dataset['date'])
```

```python
final_dataset
```

```python
#now claasify dataset into dependent and independent variables
x = dataset.iloc[:, : -1]
y = dataset.iloc[:, -1]
```

```python
#now lets split the data into training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
#now import KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors= 3)
model.fit(x_train, y_train)
```
