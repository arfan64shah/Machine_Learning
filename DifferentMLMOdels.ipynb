```python
#one hot encoding to handle the object values

dataset['dengue.p_i_d'].unique(), dataset['dengue.date_of_fever'].unique(), dataset['dengue.residence'].unique(), dataset['dengue.days'].unique()

```

```python
# we need to apply one hot encoding and columns will increase
oneHotEncoded_dataset = pd.get_dummies(dataset, columns = ['dengue.p_i_d', 'dengue.date_of_fever', 'dengue.residence', 'dengue.days'])
oneHotEncoded_dataset.head()
```

```python
#split data into dependent and independent 
x = oneHotEncoded_dataset.iloc[:, :-1]
y = oneHotEncoded_dataset.iloc[:, -1]
```

```python
#noe split the data into training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

# SVM

```python
#lets now apply SVM classifier and fit training data
from sklearn.svm import SVC
classifier = SVC()
classifier.fit(x_train, y_train)
```

```python
#as we have fitted the model, now lets predict the model
predict = classifier.predict(x_test)
```

```python
#lets find accuracy of model
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, predict)
accuracy
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predict)
cm
```
