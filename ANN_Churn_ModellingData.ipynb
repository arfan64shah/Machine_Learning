Since there only two values 0 and 1 so we will apply binary classification

```python
#as we can see in the dataset that there are many object data types. so we need to apply one hot encoding
#before that lets find how uniques are there in the object data types
print(len(dataset['Surname'].unique()))
print(dataset['Geography'].unique())
print("Gender types are ", dataset['Gender'].unique())
```

```python
#lets apply now one hot encoding
hotEncoded_dataset = pd.get_dummies(dataset, columns = ['Surname', 'Geography', 'Gender'])
hotEncoded_dataset.head()
```

As we can see that the number of columns increased 

```python
#now display all the columns
# pd.set_option('display.max_columns', None)
hotEncoded_dataset.head()
```

We can see that the target column in the middle somewhere so we need to move it to the last

```python
#move Exited column to the last
final_dataset = hotEncoded_dataset.reindex(columns = [col for col in hotEncoded_dataset.columns if col != 'Exited'] + ['Exited'])
final_dataset.head()
```

```python
#now divide the dataset in dependednt and independednt
x = final_dataset.iloc[:, :-1]
y = final_dataset.iloc[:, -1]
```

```python
#now categories training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 0)
```
