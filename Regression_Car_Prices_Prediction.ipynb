```python
#classify dataset into deoendent and independent catogries
x = final_dataset.iloc[:, :-1]
y = final_dataset.iloc[:,  -1]
```

```python
#now classify the dataset into training and testing
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)

print(len(x_train))
print(len(y_train))
print(len(x_test))
print(len(y_test))
```

```python
#apply multi regression
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()

#fit the dataset
regressor.fit(x_train, y_train)

#predict

pred = regressor.predict(x_test)
```

```python
#evaluate metrics
from sklearn.metrics import mean_absolute_error

score = mean_absolute_error(pred, y_test)

score
```

```python
#compare predicted and actaul values
reg_model_diff = pd.DataFrame({'Actual value': y_test, 'Predicted value': pred})
reg_model_diff
```

```python
#now lets validate the model
from sklearn.model_selection import cross_val_score

cvs = cross_val_score(regressor, x, y, cv = 4)
cvs = cvs

#calculate mean of both
np.mean(cvs)
```
