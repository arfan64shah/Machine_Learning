```python
#reading the csv data using pandas
import pandas as pd
BHNames= ['crim','zn','indus','chas','nox','rm',
         'age','dis','rad','tax','ptratio','black','lstat','medv']

final_dataset = pd.read_csv("housing.data", delim_whitespace=True, names=BHNames)

#dataset = pd.read_csv('estate_data.csv')
```

```python
#now lets move divide dataset into dependent and independent data
x = final_dataset.iloc[:, : -1]
y = final_dataset.iloc[:, -1]
```

```python
#now lets split data into training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
#now lets import sequential from keras library
from keras.models import Sequential
from keras.layers import Dense
classifier = Sequential()

#now lets apply first hidden layer
classifier.add(Dense(20, input_dim = 13, kernel_initializer = 'glorot_uniform', activation = 'relu'))

#add second hidden layer
classifier.add(Dense(20, input_dim = 13, kernel_initializer = 'uniform', activation = 'sigmoid'))

#add a third hidden layer
classifier.add(Dense(20, input_dim = 13, kernel_initializer = 'uniform', activation = 'relu'))
```

```python
#lets apply output layer
classifier.add(Dense(1, kernel_initializer = 'uniform', activation = 'linear'))
```

```python
#now lets compile and calculate loss function
classifier.compile(optimizer = 'Adam', loss='mean_squared_error', metrics=['accuracy'])
```

```python
classifier.fit(x_train, y_train, epochs = 1000)
```

```python
#now predicting prices
y_predict = classifier.predict(x_test)
```

```python
test_predictions = classifier.predict(x_test).flatten()
results_df = pd.DataFrame({'Predicted Price': test_predictions,'Actual Price':y_test.values.flatten()})
results_df.head(100)
```

```python
from sklearn.metrics import mean_absolute_error
print("Mean absolute error is ", mean_absolute_error(y_predict, y_test))
```
